```python
num_cols = df.select_dtypes(include=['int64', 'float64'])
for i in num_cols:
    plt.figure(figsize=(8,6))
    sns.boxplot(df[i])
```

```python
num_cols = df.select_dtypes(include=['int64', 'float64']).columns
for col in num_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    df = df[(df[col] >= lower) & (df[col] <= upper)]
df.shape
```

```python
plt.figure(figsize=(8,6))
sns.heatmap(df.corr(numeric_only=True),annot=True)
```

```python
df['Price_per_km']=df['Avg_Price_USD']/(df['Range_km']+1)
df['Age']=2025-df['Year']
df['Range_per_Year']=df['Range_km']/(df['Age']+1)
df['Average_price_per_unit']=df['Avg_Price_USD']/(df['Production_Units']+1)
df['Price_kwh']=df['Avg_Price_USD']/(df['Battery_Capacity_kWh']+1)
df['Range_per_kwh']=df['Range_km']/(df['Battery_Capacity_kWh']+1)
```

```python
x=df.drop('Avg_Price_USD',axis=1)
y=df['Avg_Price_USD']
```

```python
x_encoded=pd.get_dummies(x,columns=['Region','Model','Source_Type'],drop_first=True)
```

```python
x_encoded.head()
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
x_scaled=scaler.fit_transform(x_encoded)
```

```python
x_scaled=pd.DataFrame(x_scaled,columns=x_encoded.columns)
```

```python
x_scaled.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(x_scaled, y, test_size=0.20, random_state=42)
```

```python
from xgboost import XGBRegressor
```

```python
model=XGBRegressor(n_estimators=400,max_depth=6,learning_rate=0.05,random_state=42)
```

```python
model.fit(X_train,y_train)
```

```python
y_pred=model.predict(X_test)
```

```python
from sklearn.metrics import r2_score
```

```python
r2=r2_score(y_test,y_pred)
```
